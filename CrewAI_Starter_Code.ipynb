**CrewAI**
  * CrewAI is a Python Framework for orchestrating role-playing autonomous AI Agents that collaborate as a Crew to accomplish complex tasks.
  * CrewAI is a multi-agent framework designed to facilitate collaboration among autonomous AI agents. Built on LangChain, it aims to harness the collective intelligence of these agents to tackle complex tasks.
  * CrewAI allows agents to assume specific roles, set goals, and work together seamlessly, mimicking a well-coordinated team. This framework is particularly useful for creating multi-agent systems like smart assistant platforms, automated customer service teams, and research groups.

**Why CrewAI?**
  * Complex workflows that need multiple specialized skills — research, writing, coding, QA — are hard for a single LLM call.
  * CrewAI breaks work into agent-sized pieces with clear roles and coordination.

**CrewAI Architecture:**
  * 🧠 LLM Backend(GPT, Claude) → 🚢 Crew (Orchestrator) → 🤖 Agent (Role + Goal) → 📋 Task → 🔧 Tool (External API) → 📤 Output

**The five Main Components:**
CrewAI revolves around five main Components that work together to streamline workflows and enhance productivity. Here's an overview of each:

**1. Agents:** Agents are the autonomous entities in CrewAI designed to perform specific actions or tasks. An agent is an autonomous unit programmed to: Perform tasks, Make decisions, Communicate with other agents.

**2. Tasks:** Tasks are the individual units of work that need to be completed by Agents.

**3. Tools:** Tools are the resources and applications that Agents utilize to perform their tasks. These can include UDF, software applications, APIs, databases, and any other technological resources that aid in task completion.

**4. Process:** Process refers to the structured series of tasks and actions that lead to a specific outcome.

**5. Crew:** Crew represents the collective set of Agents, Tasks, Tools, and Processes working together towards common goals.

4. Define CrewAI Agents

**Agent**
  * Reasech Topic → Planner Agent → Researcher Agent → Writer Agent → Editor Agent → 📤 Final Research Paper.

a. Planner Agent

```python
planner_agent = Agent(
    role = "Research Planner",

    goal = """
    Analyze the research topic and create a structured planning for writing a research paper for publishing in international journal following IEEE Standard.
    Prepare a detailed planning for the following sections:
    - Title
    - Keywords
    - Abstract
    - Introduction
    - Related Work / Literature Review
    - Research Gap
    - Problem Statement
    - Research Objectives
    - Proposed Methodology
    - Experimental Setup
    - Results
    - Discussion
    - Conclusion
    - Future Work
    - References

    For every section describe:
    - Purpose
    - Key topics
    - Expected contents
    - Logical flow

    Do not write the paper.
    Produce detailed research paper blueprint.
    """,

    backstory = """
    You are an experienced academic researcher who specializes in planning high quality international research papers before writing begins.
    """,
    allow_delegation = True,
    llm = llm,
    verbose = True,
)
```

b. Researcher Agent

```python
researcher_agent = Agent(
    role = "Researcher Analyst",
    goal = """
    Coduct a comprehensive literature review based on the planner's outline.

    Identify:
    - Recent research papers
    - Existing methodologies
    - State of the art techniques
    - Research gaps
    - Public Datasets
    - Evaluation metrics
    - Performance comparisons
    - Advantages and limitations
    - Relevant citation and references

    Organize all finidings into a section-wise according to research paper planner context response.
    Do not write the research paper just create a outline and structure with context from available research papers.
    """,

    backstory = """
    You are a senior research analyst experienced in reviewing international journal publications and summarizing
    scientific literature objectively.
    """,

    llm = llm,
    allow_delegation = True,
    verbose = True,
)
```

c. Writer Agent

```python
writer_agent = Agent(
    role = "Acadamic Research Writer",
    goal = """
    Analyze the research planner and researcher context response from both agents and write the research paper.
    Generate the paper section by section following the below structutre and writing standards.

    Include :

    - Title
    - Keywords
    - Abstract
    - Introduction
    - Related Work / Literature Review
    - Research Gap
    - Problem Statement
    - Research Objectives
    - Proposed Methodology
    - Experimental Setup
    - Results
    - Discussion
    - Conclusion
    - Future Work
    - References

    Maintain a formal academic writing style and technical langauge throughout.
    """,
    backstory = """
    You are an experienced academic researcher and research paper writer who specializes in writing high quality academic research paper
    for internaltional publication.
    """,
    allow_delegation = True,
    llm = llm,
    verbose = True,
)
```

```python
editor_agent = Agent (
    role = "Research Paper Editor",
    goal = """
    Review the complete research paper before publication.

    Verify:
    - Grammer
    - Academic writing style
    - Logical Flow
    - Technical Correctness
    - Section Consistency
    - Clarity
    - Completeness
    - Citation consistency
    - Formatting
    - Publication readiness
    Improve readability while preserving the original meaning.
    Produce the final publication-ready version.
    """,
    backstory="""
    You are a senior journal editor responsible for reviewing and editing research papers according to IEEE standards.
    """,
    llm=llm,
    verbose=True,
)
```

5. Define Your Tasks

a. Planning Task

```python
planning_task = Task(
    description = """
    Analyze the research topic and create a structured planning for writing a research paper for publishing in international journal.
    Prepare a detailed blueprint for the following sections:
    - Title
    - Keywords
    - Abstract
    - Introduction
    - Related Work / Literature Review
    - Research Gap
    - Problem Statement
    - Research Objectives
    - Proposed Methodology
    - Experimental Setup
    - Results
    - Discussion
    - Conclusion
    - Future Work
    - References
  For every section provide:
    - Purpose
    - Key Topics
    - Expected Contents
    - Logical Flow""",
  expected_output = """
    A structured research paper blueprint with section-wise planning that will
    be used by the Researcher Agent.""",
  agent = planner_agent,
  input_variables = ["research_topic"],
  verbose = True,
)
```

b. Research Task

```python
research_task = Task(
        description="""
    Perform a comprehensive literature review using the research paper blueprint.

    Research and summarize:
    - Recent research papers
    - Existing methodologies
    - State-of-the-art techniques
    - Research gaps
    - Public datasets
    - Evaluation metrics
    - Performance comparison
    - Advantages
    - Limitations
    - References

    Organize the findings according to the blueprint created by the Planner Agent.
    Do not write the research paper.
    """,
    expected_output="""
    A comprehensive research report containing literature review, methodologies, datasets, references, research gaps and supporting
    information for every section of the paper.""",
    context=[planning_task],
    agent=researcher_agent,
)
```

c. Writing Task

```python
writing_task = Task(
    description="""
    Write a complete international research paper using:
    1. Research paper blueprint
    2. Literature review and research findings
    Follow academic writing standards.

    Write the following sections:
    - Title
    - Keywords
    - Abstract
    - Introduction
    - Related Work
    - Research Gap
    - Problem Statement
    - Research Objectives
    - Proposed Methodology
    - Experimental Setup
    - Results
    - Discussion
    - Conclusion
    - Future Work
    - References

    Ensure smooth transition between sections and maintain technical accuracy.
    """,
    expected_output="""
    A complete first draft of the international research paper.
    """,

    context=[
        planning_task,
        research_task
    ],
    agent=writer_agent
)
```

d. Editor Task

```python
editing_task = Task(
    description="""
    Review the complete research paper.
    Improve:
    - Grammar
    - Academic writing style
    - Clarity
    - Readability
    - Logical flow
    - Technical correctness
    - Formatting
    - Consistency
    - References
    - Publication readiness
    Produce the final publication-ready manuscript.
    """,
    expected_output="""
    Final polished research paper suitable for submission to an
    international journal according to IEEE standard.
    """,
    context=[writing_task],
    agent=editor_agent
)
```

6. Crew

```python
research_paper_crew = Crew(
    agents = [
        planner_agent,
        researcher_agent,
        writer_agent,
        editor_agent
    ],
    tasks = [
        planning_task,
        research_task,
        writing_task,
        editing_task
    ],
    process = Process.sequential,
    verbose = True
)
```

```python
result = await research_paper_crew.kickoff_async(inputs = {
    "research_topic":"AI in Agriculture Pathalogy - Deep Learning based plant disease detection using leaf images: A comparitive study of CNN and Vision Transformer Models."
    })
```
